```sql

SELECT [Date]
      ,[Ticker]
      ,[Open]
      ,[High]
      ,[Low]
      ,[Close]
      ,[Volume]
      ,[Adjusted]
      ,CASE WHEN ([DATE]= CAST('2013-01-02' AS Date)) THEN NULL
	     ELSE ROUND((LOG([Adjusted] / lag([Adjusted], 1) OVER (ORDER BY [Ticker], [Date]))),8) 
	   END AS [Return]
  FROM [TestDB].[dbo].[STOCKS]
  ORDER BY [Ticker],[Date]
```

```sql
UPDATE [TestDB].[dbo].[STOCKS] 
SET [STOCKS].[Return] = X.[Return]
FROM
(SELECT  [Date]
        ,[Ticker]
        ,CASE WHEN ([DATE]= CAST('2013-01-02' AS Date)) THEN NULL
	     ELSE ROUND((LOG([Adjusted] / lag([Adjusted], 1) OVER (ORDER BY [Ticker], [Date]))),8) 
	     END AS [Return]
  FROM [TestDB].[dbo].[STOCKS]) X
WHERE [STOCKS].[Ticker] = X.[Ticker] 
AND [STOCKS].[Date] = X.[Date];
```

```sql
USE TestDB
GO

 DECLARE @sql NVARCHAR(MAX);
 SET @sql = N'SELECT [Date] as [date]
				  ,[Ticker] 
				  ,[Adjusted] as [adjusted]
			  FROM [TestDB].[dbo].[STOCKS]
			  WHERE [Ticker] IN(''^GSPC'',''ABBV'',''ABT'')
			  ORDER BY [Ticker], Date Asc;';

DECLARE @rScript nvarchar(max)
SET @rScript = N'library(tidyquant)
                 
				 data <- SqlData;

				 Ra <- data %>%
				       filter(Ticker != ''^GSPC'') %>%
				         group_by(Ticker) %>%
				         tq_transmute(adjusted, periodReturn, period = "daily", col_rename = "Ra")

				 # Get returns for SP500 as baseline
				 Rb <- data %>%
				       filter(Ticker == ''^GSPC'') %>%
				         group_by(Ticker) %>%
				         tq_transmute(adjusted, periodReturn, period = "daily", col_rename = "Rb")

				# Merge stock returns with baseline
				RaRb <- left_join(Ra, Rb, by = c("date" = "date"))

				# Get CAPM table from PerformanceAnalytics
				capm <- RaRb %>%
					      tq_performance(Ra = Ra, Rb = Rb, performance_fun = table.CAPM)

                OutData <- capm';
EXEC sp_execute_external_script
     @language = N'R',
     @script = @rscript,
     @input_data_1 = @sql,
	 @input_data_1_name = N'SqlData',
	 @output_data_1_name = N'OutData'
WITH RESULT SETS( 
	   ([symbol] nvarchar(50), [ActivePremium] float, [Alpha] float, [AnnualizedAlpha] float,[Beta] float,
		[Beta-] float,[Beta+] float,[Correlation] float,[Correlationp-value] float,[InformationRatio] float,
		[R-squared] float, [TrackingError] float,[TreynorRatio] float)
	   ); 
GO
```
